```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Muat dataset Ford GoBike
df = pd.read_csv('fordgobike-tripdataFor201902.csv')  # Ganti dengan path dataset yang sesuai
```

```python
# Cek apakah ada missing value (data kosong)
print(df.isnull().sum())
```

```python
# Visualisasi missing values dengan heatmap
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
# Buat kopian data agar data asli tidak berubah
df_copy = df.copy()
print(df_copy.head(10))
```

```python
# Buang record yang memiliki missing value di kolom-kolom tertentu
df_copy = df_copy.dropna(how='any', subset=['start_station_name', 'end_station_name'])
```

```python
# Menampilkan 5 baris pertama dari data setelah penghapusan
df_copy.head()
```

```python
# Buang record yang missing value terlalu parah
df_copy = df_copy.dropna(how="any", subset=['start_station_id'])
df_copy = df_copy.dropna(how="any", subset=['start_station_name'])
df_copy = df_copy.dropna(how="any", subset=['end_station_id'])
df_copy = df_copy.dropna(how="any", subset=['end_station_name'])
df_copy = df_copy.dropna(how="any", subset=['member_birth_year'])
df_copy = df_copy.dropna(how="any", subset=['member_gender'])
```

```python
# Cek jumlah missing value setelah penghapusan
print(df_copy.isnull().sum())
```

```python
# Print jumlah kolom dan baris
print(f"Jumlah kolom : {df_copy.shape[1]}")
print(f"Jumlah baris : {df_copy.shape[0]}")
```

```python
# Cek kembali apakah masih ada missing value
print(df_copy.isnull().sum())
```

```python
# Tampilkan perbaikan data missing value dengan pemetaan
plt.figure(figsize=(10, 6))
sns.heatmap(df_copy.isnull(), cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Menampilkan boxplot untuk kolom 'duration_sec'
plt.figure(figsize=(15, 8))  # Ukuran lebih proporsional
plt.boxplot(df_copy["duration_sec"])
```

```python
# Menambahkan judul dan label
plt.title('Boxplot of Duration (in seconds)')
plt.ylabel('Duration (seconds)')
# Menampilkan plot
plt.show()
```

```python
# Menghitung Q1, Q3, dan IQR
Q1 = df_copy["duration_sec"].quantile(0.25)
Q3 = df_copy["duration_sec"].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas atas untuk outlier
outlierQ3 = Q3 + 1.5 * IQR

# Membuang data outlier berdasarkan persamaan 2
df_copy = df_copy[df_copy["duration_sec"] < outlierQ3]

# Menampilkan deskripsi data 'duration_sec' setelah outlier dihapus
print(df_copy["duration_sec"].describe())
```

```python
# Cek kembali deskripsi kolom 'duration_sec' setelah membuang outlier
print(df_copy["duration_sec"].describe())
```

```python
# Mengecek dimensi dataset untuk mengetahui jumlah kolom dan baris yang tersisa
print(f"Jumlah kolom data copy : {df_copy.shape[1]}")
print(f"Jumlah baris data copy : {df_copy.shape[0]}")
```

```python
# Cek kembali deskripsi kolom 'duration_sec' setelah membuang outlier
print(df_copy["duration_sec"].describe())
```

```python
# Menampilkan deskripsi statistik dari seluruh kolom
df_copy.describe(include="all")
```

```python
# Mengecek data duplikat menggunakan fungsi duplicated
duplicated_data = df_copy.duplicated()
print(f"Jumlah data duplikat: {duplicated_data.sum()}")

# Menghapus data duplikat
df_copy = df_copy.drop_duplicates()

# Mengecek kembali dimensi data setelah penghapusan data duplikat
print(f"Jumlah kolom data copy : {df_copy.shape[1]}")
print(f"Jumlah baris data copy : {df_copy.shape[0]}")
```
